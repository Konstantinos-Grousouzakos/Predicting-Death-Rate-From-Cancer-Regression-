## 2. LIBRARIES AND DATA

__Libraries__

```python
import pandas as pd
import numpy as np
from numpy import arange
import seaborn as sns
import matplotlib.pyplot as plt

import lightgbm as lgb
from sklearn.model_selection import KFold, cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

pd.set_option('display.max_columns', 100)
```

__Importing data__

```python
data1 = pd.read_csv("train_dataset.csv") 
data2 = pd.read_csv("test_dataset.csv") 

# Train and test Dataset
train_df = data1.copy()
test_df = data2.copy()
```

```python
train_df.head(2)
```

```python
test_df.head(2)
```

## 3. PREPROCESSING

```python
# Defining kfold validation
cv = KFold(n_splits=5) 
```

```python
# Defining X_train, Y_train, X_test, Y_test
X_train = train_df.drop(columns = 'TARGET_deathRate')
Y_train = train_df[ 'TARGET_deathRate']

X_test = test_df.drop(columns = 'TARGET_deathRate')
Y_test = test_df[ 'TARGET_deathRate']
```

```python
X_train.head(2)
```

```python
X_test.head(2)
```

## 4. LIGHTGBM

```python
# Defining Grid for GridSearchCV

grid = {'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [2,3,4,5,6,7],
        'num_leaves': [15, 31, 63, 127],
        'min_child_samples': [20, 40, 60,80,100, 120],
        'subsample': [0.8, 0.9, 1],
        'colsample_bytree': [0.8, 0.9, 1.0]}
```

```python
# Instantiating GridSearchCV 

LIGHTGBM_search = GridSearchCV(lgb.LGBMRegressor(boosting_type='gbdt', objective='regression', metric='mse', verbose=-1), 
                            grid, scoring='neg_mean_squared_error', cv=cv, n_jobs=-1 ,verbose = 1)
```

```python
# Searching for the best parameters 

CV_LIGHTGBM_results = LIGHTGBM_search.fit(X_train, Y_train)
```

```python
# Displaying the best parameters 

LIGHTGBM_Best_Parameters = pd.DataFrame(CV_LIGHTGBM_results.best_params_,
                                        index = ['Best LightGBM Regression Parameters'])

LIGHTGBM_Best_Parameters = LIGHTGBM_Best_Parameters.transpose()
LIGHTGBM_Best_Parameters
```

```python
# Storing the values of the best parameters
best_colsample_bytree = CV_LIGHTGBM_results.best_params_['colsample_bytree']
best_learning_rate = CV_LIGHTGBM_results.best_params_['learning_rate']
best_max_depth = CV_LIGHTGBM_results.best_params_['max_depth']
best_min_child_samples = CV_LIGHTGBM_results.best_params_['min_child_samples']
best_num_leaves = CV_LIGHTGBM_results.best_params_['num_leaves']
best_subsample = CV_LIGHTGBM_results.best_params_['subsample']
```

```python
# Instantiating the best model

LIGHTGBMModel = lgb.LGBMRegressor(boosting_type='gbdt', objective='regression', metric='mse', verbose=-1,
                                  subsample = best_subsample,
                                  learning_rate = best_learning_rate, 
                                  max_depth = best_max_depth, 
                                  colsample_bytree = best_colsample_bytree, 
                                  min_child_samples = best_min_child_samples, 
                                  num_leaves = best_num_leaves).fit(X_train, Y_train)
```

```python
# Getting the R^2

LIGHTGBM_R2 = np.round(LIGHTGBMModel.score(X_train, Y_train),3)
```

```python
# Getting predictions and training MSE 

LIGHTGBMPredictions = LIGHTGBMModel.predict(X_train)

LIGHTGBM_Training_MSE = np.round(mean_squared_error(Y_train, LIGHTGBMPredictions),2)
```

```python
# Getting kfold MSE 

lightgbm_cv_MSE = round(np.mean(cross_val_score(LIGHTGBMModel, X_train, Y_train, cv=cv, 
                                          scoring='neg_mean_squared_error', error_score='raise') * (-1)),1)
```

```python
# Getting predictions and test MSE 

LIGHTGBM_predictions = LIGHTGBMModel.predict(X_test)
LIGHTGBM_Test_MSE = round(mean_squared_error(Y_test, LIGHTGBM_predictions), 2)
```

```python
# Dictionary that holds the performance metrics 

dict1 = {'LightGBM': (np.round(LIGHTGBM_R2,3), LIGHTGBM_Training_MSE,
                                            lightgbm_cv_MSE, LIGHTGBM_Test_MSE )}
```

```python
# Creating a dataframe to easily present the results 

Evaluation_df = pd.DataFrame(dict1, index=['R^2','Training_MSE','Training_MSE (KFold)', 'Test_MSE'])
```

## 5. RESULTS

__Aggragated Results__

```python
# Displaying the metrics

Evaluation_df
```

__Residuals's Plots__

```python
residuals_lightgbm = LIGHTGBMPredictions - Y_train 

sns.set_style("white")
    
plt.figure(figsize=(5,4))
    
plt.hist(residuals_lightgbm, bins = 20, color = "forestgreen")
    
plt.title(label = "Distribution of Residuals (LightGBM)" , fontsize = 14, weight = 'bold')
    
sns.despine()
    
plt.show()
```

LightGBM demonstrates a balanced performance across the training, cross-validated, and test datasets, with training error levels comparable to those observed in cross-validation and test sets. Moreover, it consistently delivers highly accurate predictions.
